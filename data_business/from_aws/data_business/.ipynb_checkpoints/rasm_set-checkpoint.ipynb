```python
# Subset track_sm for artists in artist_all
# columns to keep = track_id, artist, genre, genre_all, listens
# keep another table for artists and their related artists?

songs_per_artist = {}
genres_in_set = {}
genres_artist = {}
genres = []
tracks_set = []

def get_reqs():
    
    global songs_per_artist, genres_in_set, genres_artist, genres, tracks_set, artist_all
    songs_per_artist = {}
    genres_in_set = {}
    genres_artist = {}
    genres = []
    tracks_set = []
    
    tracks_set = tracks_sm[tracks_sm['artist.12'].isin(artist_all.keys())]
    tracks_set = tracks_set[['artist.12', 'track.7', 'track.8', 'track.9', 'track.14']]

    # get songs per artist
    for k in artist_all:
        songs_per_artist[k] = len(tracks_set[tracks_set['artist.12'] == k])


    # get the genre information in set
    # for genres and songs per genre info
    
    for gens in tracks_set['track.9']:
        l = [re.split(']', s) for s in re.split(', |\[', gens)]
        gens = [item for sublist in l for item in sublist]
        genres.append(gens)
        for g in gens:
            if g.isdigit():
                if genres_in_set.get(int(g), -1) != -1:
                    genres_in_set[int(g)] += 1
                else:
                    genres_in_set[int(g)] = 1

    # for genres and artists per genre                
    artists_set = list(tracks_set['artist.12'])
    for i, gens in enumerate(genres):
        for g in gens:
            if g.isdigit():
                if genres_artist.get(int(g), -1) != -1:
                    genres_artist[int(g)].add(artists_set[i])
                else:
                    genres_artist[int(g)] = set()





    # checkpost - songs per genre
    # s = 0
    # for k in genres_in_set:
    #     s += genres_in_set[k]

    # print s   

```

##### Calculating D

```python

def calc_D():

    global songs_per_artist
    # total number of songs in Set
    num_songs_set = 0
    for k in songs_per_artist:
        num_songs_set += songs_per_artist[k]

    # total number of artists in Set
    num_artists_set = len(songs_per_artist.keys())

    # std dev of songs per artist
    data = [songs_per_artist[k] for k in songs_per_artist]
    data = np.array(data)
    mn = np.mean(data, axis=0)
    sd = np.std(data, axis=0)

    D = num_songs_set/(num_artists_set * sd / num_songs_set) # formulation number 1

#     mn, sd, D
    return D
```

##### Calculating C

```python
def calc_C():

    global genres_in_set, songs_per_artist, genres_artist
    # total number of genres in Set
    num_gen_set = len(genres_in_set)

    # total number of artists in Set
    num_artists_set = len(songs_per_artist.keys())

    # std dev of artists per genre
    data = np.array([len(genres_artist[k]) for k in genres_artist])
    mn = np.mean(data, axis=0)
    sd = np.mean(data, axis=0)

    C = num_gen_set/(num_artists_set * sd / num_gen_set)

#     mn, sd, C
    return C
```

##### Calculating S

```python
def calc_S():

    global artist_rel_subset, songs_per_artist, artist_all
    # total number of edges -- change when changing the Set
    artists_in_set = artist_rel_subset

    num_edges = 0
    for key in artists_in_set:
        for elem in artists_in_set[key]:
            if elem in artists_in_set:
                num_edges += 1

    # total number of artists in Set
    num_artists_set = len(songs_per_artist.keys())

    # std dev of artist connections per artist

    data = np.array([len(artist_all[k]) for k in artist_all])
    mn = np.mean(data, axis=0)
    sd = np.mean(data, axis=0)

    S = num_edges/(num_artists_set * sd / num_edges)

#     mn, sd, S
    return S
```
