```python
%matplotlib inline

import pickle
from IPython.display import display, HTML
import matplotlib.pyplot as plt
```

```python
# saving the artist_rel table

def save_artists(name, table):
    with open('%s.pickle'%name, 'wb') as writer:
        pickle.dump(table, writer, protocol=pickle.HIGHEST_PROTOCOL)

    # loading artist_rel table
    # with open('artist_rel.pickle', 'rb') as handle:
    #     artist_rel = pickle.load(handle)
    
    
def load_things(name):
    with open(name, 'rb') as handle:
        thing = pickle.load(handle)
    
    return thing
```

```python
# Individual Components - mean, max

import seaborn as sns
import pandas as pd
import numpy as np

data = load_things("saved_data/overall_GCSD_small_train_20.pickle")
d1 = [[x[1] for x in x_d] for x_d in data]
d2 = [[x[2] for x in x_d] for x_d in data]
d3 = [[x[3] for x in x_d] for x_d in data]
d4 = [[x[4] for x in x_d] for x_d in data]
d1 = np.array(d1)
# d1 = np.mean(d1, axis=0)
d1 = np.max(d1, axis=0)
d2 = np.array(d2)
# d2 = np.mean(d2, axis=0)
d2 = np.max(d2, axis=0)
d3 = np.array(d3)
# d3 = np.mean(d3, axis=0)
d3 = np.max(d3, axis=0)
d4 = np.array(d4)
# d4 = np.mean(d4, axis=0)
d4 = np.max(d4, axis=0)

x = range(50, 390, 10)
df = pd.DataFrame({'x': x, 'Goodness': d1, 'C': d2, 'D': d3, 'S': d4})
# style
plt.style.use('seaborn-bright')
 
# create a color palette
palette = plt.get_cmap('Set1')
 
# multiple line plot
num=0
for column in df.drop('x', axis=1):
    num+=1
    plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)
 
# Add legend
plt.legend()
 
# Add titles
plt.title("Max GCSD v N for 50 iterations (Small Training)", loc='left', fontsize=12, fontweight=0, color='black')
plt.xlabel("N")
plt.ylabel("Max Goodness")
plt.savefig("figures/GCSD_maxvN_small_training_20.png")
```
