```python
# saving the artist_rel table

def save_artists(name, table):
    with open('%s.pickle'%name, 'wb') as writer:
        pickle.dump(table, writer, protocol=pickle.HIGHEST_PROTOCOL)

    # loading artist_rel table
    # with open('artist_rel.pickle', 'rb') as handle:
    #     artist_rel = pickle.load(handle)
    
    
def load_things(name):
    with open(name, 'rb') as handle:
        thing = pickle.load(handle)
    
    return thing
```

```python
# RUn TIme Plots
import seaborn as sns
import pandas as pd
import numpy as np


the_times = load_things("saved_data/the_times_small_train_2.pickle")
```

```python
n = [x_d[1] for x_d in the_times]
calls = [x_d[0] for x_d in the_times]
times = [x_d[2] for x_d in the_times]
times2 = [times[0]]
times3 = [times[0]]

calls2 = np.multiply(calls, n)
for i in range(1, len(calls)):
    t = (calls2[i] * times2[i-1])/calls2[i-1]
    times2.append(t)
    t = (calls[i] * times3[i-1])/calls[i-1]
    times3.append(t)



```

```python
times3
```

```python


df = pd.DataFrame({'x': n, 'actual runtime': times, 'O(N*N*(M-N))': times2, 'O(N*(M-N))': times3})
plt.style.use('seaborn-bright')
 
# create a color palette
palette = plt.get_cmap('Set1')

num=0
for column in df.drop('x', axis=1):
    num+=1
    plt.plot(df['x'], df[column], marker='', color=palette(num), linewidth=1, alpha=0.9, label=column)
 

# Add legend
plt.legend()

plt.title("Run Time v N (Small Training)", loc='left', fontsize=12, fontweight=0, color='black')
plt.xlabel("N")
plt.ylabel("Run Time (sec)")
plt.savefig("figures/runtimevN_2.png")
```
